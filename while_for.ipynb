```python
prateleira= ['banana','maça','laranja','uva']

#o comando len(variavel) mosta a quantidade de elementos dentro de algo
print(len(prateleira))

# criar um loop que pergunte ao usuario se ele quer adicionar uma fruta na prateleira
# a prateleira so pode ter 7 itens
# sempre que adicionar um produto deve mostrar seus valores



while len(prateleira) <=7:
  pergunta = input('deseja adicionar mais um item à prateleia ')
  print(f'itens na prateleira {prateleira}')

  if pergunta.lower() == 'sim':
    nova_fruta = input('adicione o item: ')
    prateleira.append(nova_fruta)
    print(f'item adicionado: {prateleira}')

    for i in prateleira: #o *for* deixa de exibir um por um, de forma extensa, e comaça a exibir em forma de lista
      print(i) #o "i" é só um apelido temporário que você dá pra cada item da lista enquanto o loop passa por eles.


      #um outro exemplo
  #for i in prateleira
    #print (f'temos a fruta: {i}')


  else:
    print(f'itens da prateleira {prateleira}')
```

```python
#FOR para uma lista

lista = ['aA','bB','cB']

for i in lista:
  print(i)

#FOR para contagem
#o range sempre começa no 0
for i in range(10):
  print(i+1) # i+1 para que inicie no 1, se quisesse que iniciasse no 0, e terminasse no 10, atribuiria ao range o numero 11
```
