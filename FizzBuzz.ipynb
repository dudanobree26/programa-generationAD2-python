```python
# fizzbuzz é um jogo no qual consiste em;
# sempre que o número é divisivel por 3 mostra FIZZ
# sempre que é divisivel por 5 monstra BUZZ
# quando é divisivel por 3 e por 5 monstra FIZZBUZZ
# caso nao divisivel por nenhum dos numeros sitados, ele nao fdaz nada

numero = int(input('informe um numero: '))

if numero < 100:
  print('numero muito baixo') # se acaso o numero for <100 automaticamente o numero sera alterado para 100, o menor numero possivel para o jogo
  numero = 100
else:
  print('sistema pronto para iniciar')

for i in range (numero):
  #print(i+1)
  if (i+1) % 3 == 0 and (i+1) % 5 == 0:
    print(f'o numero {i+1} é fizzbuzz')
  elif (i+1) % 3 == 0:
    print(f'numero {i+1} é fizz')
  elif (i+1) % 5 == 0:
    print(f'numero {i+1} é buzz')
  else:
    print(f'o numero {i+1} não é nada')

```
