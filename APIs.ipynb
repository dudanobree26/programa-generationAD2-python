```python
import requests
#vamos usar a blibioteca requets para ver o codigo do status api
#status code
#200--esta funcionando | 404-- arquivo não encontrado

#toda API tem uma propiedade chamada de end-point

url = 'https://api.open-meteo.com/v1/forecast'

#todas API precisa de parametros para funcionar
#precisamos definir os parametros para uso
latitude= float(input('informe a latitude desejada: '))
longitude= float(input('informe a longitude desejada: '))

parametros={
    'latitude': latitude,
    'longitude': longitude,
    'current_weather': True
}
#vamos fazer uma requisição para coletar dados do sistema
#não posso so pegar (get) a informação
#temos que registrar
resposta = requests.get(url,params=parametros)
#pra coletar algo, chamamos a blibioteca requests com o atriuto get (onde, oque)
# request.get(url,parans=parametros)

#vamos verificas se o API esta funcionando
if resposta.status_code == 200:
  #vamos fazer funcionar
  #API é melhor que json
  #vamos colocar a resposa em um jason
  dados = resposta.json()

# vamos falar o que queremos do api
  temperatura = dados['current_weather']['temperature']
  velocidade_vento = dados['current_weather']['windspeed']

  print(f'a temperatura é {temperatura}°C')
  print(f'à velocidade do vento é {velocidade_vento}km/h')

# se a api caiu ou não funciona
else:
  print(f'erro ao acessar a API - {resposta.status_code}')
```
