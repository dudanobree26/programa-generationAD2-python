```python
#para conseguir usar a bliblioteca precisamos importar ela
import pandas as pd # <- IMPORT nome_da_bliblioteca AS apelido_para_uso
```

```python
#esqueleto do arquivo
#vamos usar dicionarios oara criar os dados
#dicionarios funciona com a aideia de chave->valor ou chave->valores

dados = {
    'nomes':['andre','paulo','aline','paola'],
    'idade':[36,25,30,18],
    'cidade':['Sorocaba','indaiatuba','sao paulo','registro']
}
#a pandas pode tranformar um dicionario em uma "tabela"
#esse pedaço de tabela é chamado de dataframe

#para tranformar
df_dados = pd.DataFrame(dados)

#print(df_dados)

display(df_dados)
```

```python
#ciar um arquivo csv (excel)

#para criar , precisamos de dados
#chamamos os dados e atibuimos uma funcao -> "to_------" (nome, index)

df_dados.to_csv("teste_dados.csv",index= False)
```

```python
#como ler um arquivo que me enviaram

df_ler_dados = pd.read_csv('/content/teste_dados.csv')

#comando para ler as primeiras infos do arquivo
df_ler_dados.head()
```

```python
# vamos popular a tabela

novos_dados = {
    'nome':['Flávia','Cláudio','Fernanda'],
    'idade':[50,63,32],                      #isso é uma blibioteca
    'cidade':['Sorocaba','Iguape','Curitiba']
}
df_novos_dados = pd.DataFrame(novos_dados)
# sempre que eu quero trabalhar com dados, preciso fazer ser dados
# agora temos que juntar o antigo com o novo
# juntar = concatenar
# cria a base que vai receber todos os dados
# usa a função pd.concat para 'juntar' os dados
# pd.concat([dados_antigos, dados_novos], ignore_index=True)
df_dados_juntos = pd.concat([df_ler_dados, df_novos_dados], ignore_index=True)

# voltar para o csv
df_dados_juntos.to_csv('teste_dados.csv', index=False)
display(df_dados_juntos)
```
