```python
#usando o numpy
#matematica basica

numero1 = float(input('informe o primeiro número'))
numero2 = float(input('informe o segundo número'))
#numero2 =int(input('informe o segundo número'))
    # se eu quisesse usar o SUM


#soma pode ser sum ou add
print(f'soma:{np.add(numero1, numero2)}')
#ou o metodo sum - so para numeros inteiros (int)
#print(f'soma:{np.sum(numero1, numero2)}')

#subtrair
print(f'subtração; {np.subtract(numero1, numero2)}')
#multiplicar
print(f'multiplicação; {np.multiply(numero1, numero2)}')
#dividir
print(f'divisão; {np.divide(numero1, numero2)}')
```

```python
while True:
  numero_1 = float(input('informe o primeiro número'))
  numero_2 = float(input('informe o segundo número'))
  # while True nao necessita de uma condição para
  # inicio ou fim
  # ------
  # para finalizar o while true usamos o caso BREAK
  print('---- CALCULADORA ----')
  print('1 - somar')
  print('2 - subtrair')
  print('3 - multiplicar')
  print('4 - dividir')
  print('0 - sair')

  op = input('escolha a opção')

  if op == "0":
    break
  elif op == "1":
    print(f'soma:{np.add(numero_1, numero_2)}')
  elif op == "2":
    print(f'subtração; {np.subtract(numero_1, numero_2)}')
  elif op == "3":
    print(f'multiplicação; {np.multiply(numero_1, numero_2)}')
  elif op == "4":
    print(f'divisão; {np.divide(numero_1, numero_2)}')
  else:
    print('opção não encontrada')
```

```python
# usando sistema de calculadora - CAIXA DE SUPERMERCADO
# quais elementos tem o pagamento?

# temos produtos
# temos quantidades
# temos um total
lista_nome = []
lista_quantidade = []
lista_valor = []
while True:

  op_tem = input('Tem algum produto para passar? S/N ')
  if op_tem.lower() == 'n':
    print('sistema finalizado ')
    break
  elif op_tem.lower() == 's':
    while True:
      # precisamos de variaveis
      # precisamos de listas

      nome = input('Qual o produto? ')
      valor = float(input('Qual o valor? '))
      quantidade = float(input('Qual a quatidade? '))
      lista_nome.append(nome)
      lista_valor.append(valor)
      lista_quantidade.append(quantidade)

      op_novo_produto = input('tem outro produto? S/N ')

      if op_novo_produto.lower() == 'n':
        print('inprimindo a nota fical...')
        # vai ter coisas aqui
        print('---------- CUMPOM FISCAL ----------')
        print('Produto | Valor | Quantidade | Total do item')
        # i - produto --- j - valor ----- k - quantidade ---- l - total
        # o total é uma multiplicação - valor e a quantidade
        # transformar a lista em numpy
        np_lvalor = np.array(lista_valor)
        np_quantidade = np.array(lista_quantidade)
        # np.array() - transforma um conjunto de numeros para numpy

        total = np.multiply(np_lvalor, np_quantidade)
        valor = np.multiply(total, total)
        for i,j,k,l in zip(lista_nome, lista_valor, lista_quantidade, total):
          print(f'{i} -- {j} -- {k} -- {l}')

        break
      else:
        print('informe o novo produto')
```
